# Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
# Adjusted path to match local structure
df = pd.read_excel("../data/High_Quality_Diet_Data.xlsx")

print("Shape:", df.shape)
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Target distribution
plt.figure(figsize=(6,4))
df["Diet_Category"].value_counts().plot(kind="bar")
plt.title("Diet Category Distribution")
plt.xlabel("Diet Category")
plt.ylabel("Count")
plt.show()
```

```python
# Age group distribution
df["Age_Group"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(6,6))
plt.title("Age Group Distribution")
plt.ylabel("")
plt.show()
```

```python
# Correlation (numeric only)
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Numeric Feature Correlation")
plt.show()
```
