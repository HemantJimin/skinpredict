# Diet Recommendation & Treatment Logic

```python
def recommend_foods(skin_problem):
    if "Acne" in skin_problem:
        return ["Fruits", "Green vegetables", "Zinc-rich foods"], ["Junk food", "Sugar"]
    if "Dry" in skin_problem:
        return ["Nuts", "Seeds", "Healthy oils"], ["Alcohol", "Caffeine"]
    return ["Balanced diet"], ["Excess sugar"]

def treatment_method(trust_ai, ayurveda_awareness):
    if ayurveda_awareness >= 4:
        return "Ayurvedic"
    if trust_ai >= 4:
        return "AI-assisted"
    return "Modern / Chemical"

# Example
foods, avoid = recommend_foods("Acne-prone skin")
method = treatment_method(trust_ai=5, ayurveda_awareness=3)

foods, avoid, method
```
